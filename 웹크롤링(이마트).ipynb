### 페이지 넘어가며 웹크롤링

```python
driver = webdriver.Chrome('chromedriver')
driver.get('http://emart.ssg.com')

driver.find_element_by_xpath('//*[@id="e_gnb"]/div/div[1]/div[2]/ul[1]/li[9]/a').click()

driver.implicitly_wait(10)

product_name_list = []
price_list = []

for j in range(1,4):
    for i in range(1, 81):
        try:
            product_name = driver.find_element_by_xpath('//*[@id="ty_thmb_view"]/ul/li['+str(i)+']/div[2]/div[2]/div/a/em[1]').text
            price = driver.find_element_by_xpath('//*[@id="ty_thmb_view"]/ul/li['+str(i)+']/div[2]/div[3]/div/em').text
            image_url = driver.find_element_by_xpath('//*[@id="ty_thmb_view"]/ul/li['+str(i)+']/div[1]/div[2]/a/img[1]').get_attribute('src')
            product_name_list.append(product_name)
            price_list.append(price)
        
        except Exception as e:
            print(e)
        
        finally:
            pass
            
    
    driver.find_element_by_xpath('//*[@id="area_itemlist"]/div[2]/a['+str(j)+']').click()
    print(j, 'Done')
    
driver.quit()
```

```python
pd.DataFrame({'제품명':product_name_list,
             '가격':price_list})
```
